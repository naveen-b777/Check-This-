## Importing artist name and price from the website https://www.mojarto.com/artworks/painting/c 

```python
import urllib.request as url
from bs4 import BeautifulSoup as bs
import pandas as pd

d= url.urlopen('https://www.mojarto.com/artworks/painting/c').read().decode()
soup= bs(d,'html.parser',multi_valued_attributes=None)                           

p1= soup('ul')
p2= soup.find_all('a',class_='productTitle')

lis_p2= []
#for i in p2:
    #print(i.text.strip())

for i in p2:
    lis_p2.append(i.text.strip())
#print(lis_p2)

p3= soup.find_all('p',class_='price')
#for i in p3:
    #print(i.text.strip())

lis_p3=[]   
for i in p3:
    lis_p3.append(i.text.strip())
#print(lis_p3)

lis_p4=[]
p4= soup.find_all('p',class_='productTitle')
for i in p4:
    lis_p4.append(i.text.strip())
#print(lis_p4)


d1= {lis_p2[i]:lis_p3[i] for i in range(len(lis_p2))}
#print(d1)


df= pd.DataFrame({'Artist':lis_p2,'Art':lis_p4,'Price':lis_p3})
df
```
